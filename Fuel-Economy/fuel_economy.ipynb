#### Droping the extraneous columns
the ones that aren't consistent (not present in both datasets) or aren't relevant to our questions  
For the 2008 `"Stnd", "Underhood ID", "FE Calc Appr", "Unadj Cmb MPG"`   
For the 2018 `"Stnd", "Stnd Description", "Underhood ID", "Comb CO2"`

````python
# take a copy of the datasets to apply the changes, and to keep the original datasets unchanged
fuel_08 = fuel_08_full.copy()
fuel_18 = fuel_18_full.copy()

# the names of the columns to be dropped
columns_08 = ["Stnd", "Underhood ID", "FE Calc Appr", "Unadj Cmb MPG"]
columns_18 = ["Stnd", "Stnd Description", "Underhood ID", "Comb CO2"]

# removing the columns 
fuel_08.drop(columns=columns_08 , inplace=True)
fuel_18.drop(columns=columns_18 , inplace=True)
````

````python
fuel_08.head()
````

#### Rename columns for consistency

Rename "Sales Area" in 2008 to  "Cert Region" as in 2018

````python
fuel_08.rename(columns={"Sales Area": "Cert Region"}, inplace=True)
````

#### Rename columns to replace spaces with underscores

````python
# replace spaces with '_' in the column names
fuel_08.columns = fuel_08.columns.str.replace(' ', '_')
fuel_18.columns = fuel_18.columns.str.replace(' ', '_')

# change column names to the lowecase
fuel_08.columns = fuel_08.columns.str.lower()
fuel_18.columns = fuel_18.columns.str.lower()
````

#### Filter Dataset
We can filter the dataset to only select cars certified by California standards.  
Filter both datasets using query to select only rows where cert_region is CA. Then, drop the cert_region columns, since it will no longer provide any useful information (we'll know every value is 'CA').

````python
# for the 2008 dataset
fuel_08_CA = fuel_08.query('cert_region == "CA"').copy()
fuel_08_CA.drop(columns="cert_region", inplace=True)

# for the 2018 dataset
fuel_18_CA = fuel_18.query('cert_region == "CA"').copy()
fuel_18_CA.drop(columns="cert_region", inplace=True)

print(fuel_08_CA.shape)
print(fuel_18_CA.shape)
````

#### Drop nulls

````python
fuel_08_CA.dropna(axis=0, inplace=True)
fuel_18_CA.dropna(axis=0, inplace=True)

#check to confirm last step
fuel_18_CA.info()
````

##### Fix `air_pollution_score` datatype  
+ 2008: extract int from string.
+ 2018: convert float to int.

The following line of code raised an error `ValueError: could not convert string to float: '6/4'`
```      
# convert 2008 air_pollution_score column to float
fuel_08_CA['air_pollution_score'] = fuel_08_CA['air_pollution_score'].astype('float64')
```

````python
fuel_08_CA[fuel_08_CA.air_pollution_score == '6/4']
````

As can be seen `air_pollution_score` has value `6/4` and also `city_mpg`, `hwy_mpg`, `cmb_mpg`, `greenhouse_gas_score` 
maybe that's why these were all saved as strings! According to [this link](http://www.fueleconomy.gov/feg/findacarhelp.shtml#airPollutionScore), from the PDF documentation:

"If a vehicle can operate on more than one type of fuel, an estimate is provided for each fuel type."  
  
So all vehicles with more than one fuel type, or hybrids, like the one above (it uses ethanol AND gas) will have a two values separated by '/'- one for each type of fuel.

````python
# get all the hybrids fuel in 2008
hb_08 = fuel_08_CA[fuel_08_CA['fuel'].str.contains('/')]
hb_08
````

````python
# hybrids fuel in 2018
hb_18 = fuel_18_CA[fuel_18_CA['fuel'].str.contains('/')]
hb_18
````

I am going to take each hybrid row and split them into two new rows - one with values for the first fuel type (values before the "/"), and the other with values for the second fuel type (values after the "/"). Let's separate them with two dataframes!

##### For the 2008 

````python
# create two copies of the 2008 hybrids dataframe
df1 = hb_08.copy()  # data on first fuel type of each hybrid vehicle
df2 = hb_08.copy()  # data on second fuel type of each hybrid vehicle
````

````python
# columns to split by "/"
split_columns = ['fuel', 'air_pollution_score', 'city_mpg', 'hwy_mpg', 'cmb_mpg', 'greenhouse_gas_score']

# apply split function to each column of each dataframe copy
for c in split_columns:
    df1[c] = df1[c].apply(lambda x: x.split("/")[0])
    df2[c] = df2[c].apply(lambda x: x.split("/")[1])
````

````python
# this dataframe holds info for the FIRST fuel type of the hybrid
# aka the values before the "/"s
df1
````

````python
# this dataframe holds info for the SECOND fuel type of the hybrid
# aka the values after the "/"s
df2
````

````python
# combine dataframes to add to the original dataframe
new_rows = df1.append(df2)

# check to confirm the last step
new_rows
````

````python
# drop the original hybrid rows
fuel_08_CA.drop(hb_08.index, inplace=True)

# add the newly separated rows
fuel_08_CA = fuel_08_CA.append(new_rows, ignore_index=True)
````

````python
# check that all the original hybrid rows with "/"s are gone
fuel_08_CA[fuel_08_CA['fuel'].str.contains('/')]
````
